```python
# SPLITTING A STRING INTO MULTIPLE COLUMNS:
import pandas as pd
df = pd.DataFrame({'Name': ['Alberto Franco', 'Gino Mcneill', 'Ryan Parkes', 'Eesha Hinton', 'Syed Wharton'],
'Location': ['Melbourne, Australia', ' los angeles ,america', 'derby ,australia', 'Melbourne, Australia', 'Melbourne, Australia'],}) # create dataframe and stored in df variable
df
```

```python
df.Name.str.split(' ') # split Name column into multiple columns and stored in df variable
```

```python

df.Name.str.split(' ', expand=True) # split Name column into multiple columns and stored in df variable
```

```python
# ADDING THOSE SPLITS INTO NEW COLUMNS:

df[['First_Name', 'Last_Name']] = df.Name.str.split(' ', expand=True) # split Name column into multiple columns and stored in df variable
```

```python
df
```

```python
df[['City', 'Country']] = df.Location.str.split(',', expand=True) # split Location column into multiple columns and stored in df variable
df
```

```python
# refine data manipulation

df = df[['First_Name', 'Last_Name', 'City', 'Country']]
df
```

```python
# RESHAPING A DATAFRAME:
fasla = pd.DataFrame([['12345',100,200,300],['34567',400,500,600],['67890',700,800,900]],
                     columns=['zip','factory','warehouse','retail']) # create a new dataframe with 3 rows and 4 columns and stored in fasla
fasla.head() # show first 5 rows of fasla
```

```python
fasla.head().T # transpose the dataframe and show first 5 rows of fasla
```

```python
fasla2 = pd.DataFrame([[1,'12345','factory'], [2,'34567','warehouse'], [3,'67890','retail']], 
                      columns=['id','zip','location']) # create a new dataframe with 3 rows and 3 columns and stored in fasla2
fasla2.head() # show first 5 rows of fasla2
```

```python
fasla_long=fasla.melt(id_vars='zip', var_name='location', value_name='distance') # melt means reshape the dataframe and show first 5 rows of fasla
fasla_long.head()
```

```python
import seaborn as sns
sns.barplot(x='zip', y='distance', hue='location', data=fasla_long) # plot barplot and show first 5 rows of fasla
```
