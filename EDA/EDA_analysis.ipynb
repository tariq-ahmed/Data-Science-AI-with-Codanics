```python
# lets download the titatinc dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset from seaborn library
df = sns.load_dataset('titanic') # titanic dataset is available in seaborn library itself as well as in pandas library and stored as df dataframe.
```

```python
#  find the mean fare price of both male and female using groupby function
df.groupby('sex')['fare'].mean() # mean fare price of both male and female using groupby function
```

```python
df.groupby('class')['fare'].mean() # mean fare price of three classes using groupby function
```

```python
#  find two columns using groupby function
df.groupby(['survived', 'sex'])['fare'].mean() # find two columns using groupby function
```

```python
df.groupby(['survived', 'class' , 'who'])['fare'].mean() # find three columns using groupby function  
```

```python
df.groupby(['survived', 'who']).size() # count of survived and who using groupby function
```

```python
# correlation matrix only for numerical columns
correlation_df = df[['fare', 'age', 'sibsp', 'parch']].corr() # correlation matrix only for numerical columns
```

```python
sns.heatmap(correlation_df, annot = True) # heatmap of correlation matrix only for numerical columns
```
