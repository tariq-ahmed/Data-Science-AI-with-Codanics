# Feature Encoding

Feature encoding is the process of transforming `categorical features` into `numeric features`. This is necessary because machine learning algorithms can only handle numeric features. There are many different ways to encode categorical features, and each method has its own advantages and disadvantages. In this notebook, we will explore some of the most popular methods for encoding categorical features, such as:

- Label encoding
- Ordinal encoding
- One-hot encoding
- Binary encoding

## Assignment: how many types of feature encoding are there?

There are several types of feature encoding techniques commonly used in data preprocessing for machine learning. Here are some of the most widely used types:

1. **Label Encoding**: Converts categorical labels into integer values.
2. **One-Hot Encoding**: Converts categorical labels into binary vectors.
3. **Binary Encoding**: Converts categories into binary code and then splits the binary code into separate columns.
4. **Target Encoding**: Replaces a categorical value with the mean of the target variable for that category.
5. **Frequency Encoding**: Replaces categories with their frequency of occurrence.
6. **Hash Encoding**: Uses a hash function to convert categories into a fixed number of columns.
7. **Ordinal Encoding**: Assigns ordinal values to categories based on their order.

Each encoding technique has its specific use cases and is chosen based on the nature of the data and the requirements of the machine learning model.

## Assignment: When to use which type of feature encoding?




Feature encoding is a crucial step in preparing data for machine learning models. The type of encoding you use depends on the nature of the data and the requirements of the model. Here are some common types of feature encoding and when to use them:

1. **Label Encoding**:
   - **Use Case**: When the categorical variable is ordinal (i.e., there is a meaningful order between the categories).
   - **Example**: Encoding categories like "low", "medium", and "high" into 0, 1, and 2.

2. **One-Hot Encoding**:
   - **Use Case**: When the categorical variable is nominal (i.e., there is no meaningful order between the categories).
   - **Example**: Encoding categories like "red", "blue", and "green" into binary vectors.

3. **Binary Encoding**:
   - **Use Case**: When you have high cardinality categorical variables (i.e., variables with many unique categories) and want to reduce the dimensionality.
   - **Example**: Encoding a category with many unique values into a binary representation.

4. **Target Encoding**:
   - **Use Case**: When you want to encode categorical variables based on the target variable. This is often used in scenarios where the relationship between the categorical variable and the target variable is important.
   - **Example**: Encoding categories based on the mean of the target variable for each category.

5. **Frequency Encoding**:
   - **Use Case**: When you want to encode categories based on their frequency of occurrence.
   - **Example**: Encoding categories based on how often they appear in the dataset.

6. **Hash Encoding**:
   - **Use Case**: When you have high cardinality categorical variables and want to reduce the dimensionality without creating a large number of columns.
   - **Example**: Using a hash function to encode categories into a fixed number of columns.

### Example Code for One-Hot Encoding in Python using Pandas



```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'blue', 'red']
})

# One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=['color'])
print(df_encoded)
```



### Example Code for Label Encoding in Python using Scikit-Learn



```python
from sklearn.preprocessing import LabelEncoder

# Sample DataFrame
df = pd.DataFrame({
    'size': ['small', 'medium', 'large', 'medium', 'small']
})

# Label Encoding
label_encoder = LabelEncoder()
df['size_encoded'] = label_encoder.fit_transform(df['size'])
print(df)
```
