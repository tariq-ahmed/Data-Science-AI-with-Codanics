# Seabonr and Plotting

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Important things for plotting according to grammer of graphics


1. Data (df)
2. Types of plot (scatter, line, bar, etc)
3. x axis (x)
4. y axis (y)
5. Grouping (hue/color) based on categorical variable
6. Size of plots (size)
7. Color palette for plots (palette)
8. Style of plots (style)
9. How do you interpret the plot (title, legend, etc)

```python
# Data
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
# scatter plot
# scatter plot size of figure , A4: 8.27 x 11.69, slide size 16:9
sns.scatterplot(data= df, x='age', y='fare', hue='sex')
```

```python
# scatter plot
# scatter plot size of figure , A4: 8.27 x 11.69, slide size 16:9
plt.figure(figsize=(16,9))
sns.scatterplot(data= df, x='age', y='fare', hue='sex')
```

```python
# scatter plot

# scatter plot size of figure , A4: 8.27 x 11.69, slide size 16:9
plt.figure(figsize=(16,9))

# change the scatter point colors
pallet={'male':'blue', 'female': 'red'} 
sns.scatterplot(data= df, x='age', y='fare', hue='sex' , palette=pallet)

# labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')


# saving a plot 300 dpi or 600 dpi, 150 dpi is good for web and save as png
plt.savefig('./outputs/plots/scatter_plot_zz.png', dpi=300)

# saving a plot in vector format svg
plt.savefig('./outputs/plots/scatter_plot_zz.svg')
```

```python
# Style of Plot
sns.set_theme(style='whitegrid')

# scatter plot size of figure
plt.figure(figsize=(8,4))

# change the scatter point colors
pallet={'male':'blue', 'female': 'red'} 
sns.scatterplot(data= df, x='age', y='fare', hue='sex' , palette=pallet)

# labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
```

```python
# Style of Plot
sns.set_theme(style='darkgrid')

# scatter plot size of figure
plt.figure(figsize=(8,4))

# change the scatter point colors
pallet={'male':'blue', 'female': 'red'} 
sns.scatterplot(data= df, x='age', y='fare', hue='sex' , palette=pallet)

# labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
```

```python
# Font Scale of Plot and font name

# Style of Plot
sns.set_theme(style='darkgrid' , font='Times New Roman', font_scale=1.5)

# scatter plot size of figure
plt.figure(figsize=(16,9))

# change the scatter point colors
pallet={'male':'blue', 'female': 'red'} 

# size of scatter points according to age of passenger increase
sns.scatterplot(data= df, x='age', y='fare', hue='sex' , size='age', palette=pallet)

# labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
```

Interpertation:

1. We have age of passengers at x-axis and fare at y-axis
2. The scatter plot shows the relantionship between age and fare of titanic passengers.
3. We further grouped the passengers by genders such as male (Green points) and female (Red points).
4. Two male passengers gave the highest fare in titanic, and they might be travelling to first class.
5. The oldest passenger at titanic is also male having age of 80 years old in the scatter plot. (As shown in Figure 1)

# Boxplot

```python
df['class'].unique()
```

```python
# Style of Plot
sns.set_theme(style='darkgrid' , font='Times New Roman', font_scale=1)

# Boxplot size of figure
plt.figure(figsize=(12,6))

# change the  Boxplot colors
pallet={'First':'blue', 'Second': 'red', 'Third': 'green'}

sns.boxplot(data= df, x='sex', y='age', hue='class' , palette=pallet)

# labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxplot Plot of Age, based on Gender and Class Name of Passenger')

# change legend position and title
plt.legend(loc='upper left', title='Class Name')
```

# Boxenplot

```python
# Style of Plot
sns.set_theme(style='darkgrid' , font='Times New Roman', font_scale=1)

# Boxenplot size of figure
plt.figure(figsize=(12,6))

# change the Boxenplot colors
pallet={'First':'green', 'Second': 'red', 'Third': 'blue'}

sns.boxenplot(data= df, x='sex', y='age', hue='class' , palette=pallet)

# labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxen plot Plot of Age, based on Gender and Class Name of Passenger')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks (ticks=[0,1], labels=['Male', 'Female'], rotation=45)
```

# Barplot

```python
# Bar Plot using seaborn

# Style of Plot
sns.set_theme(style='darkgrid' , font='Times New Roman', font_scale=1)

# Boxenplot size of figure
plt.figure(figsize=(12,6))

# change the Boxenplot colors
pallet={'First':'green', 'Second': 'red', 'Third': 'blue'}

sns.barplot(data= df, x='sex', y='age', hue='class' , palette=pallet, capsize=0.1) # capsize=0.1 is for error bar

# labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Bar plot Plot of Age, based on Gender and Class Name of Passenger')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks (ticks=[0,1], labels=['Male', 'Female'], rotation=45)
```

barplot shows the mean of the data and the error bar shows the standard deviation of the data.

# Violinplot

```python
# violin Plot using seaborn

# Style of Plot
sns.set_theme(style='darkgrid' , font='Times New Roman', font_scale=1)

# violin plot size of figure
plt.figure(figsize=(12,6))

# change the violin plot colors
pallet={'First':'green', 'Second': 'red', 'Third': 'blue'}

sns.violinplot(data= df, x='sex', y='age', hue='class' , palette=pallet)

# labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Violin Plot of Age, based on Gender and Class Name of Passenger')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks (ticks=[0,1], labels=['Male', 'Female'], rotation=45)
```
