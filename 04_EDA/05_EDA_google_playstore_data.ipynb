```python
# Convert the whole Size column into bytes

# let's define a function

def convert_size(size):
    if isinstance(size,str):
        if 'k' in size:
            return float(size.replace('k',''))*1024
        elif 'M' in size:
            return float(size.replace('M',''))*1024*1024
        elif 'Varies with device' in size:
            return np.nan
    return size
```

```python
df['Size']
```

```python
# let's apply the function on the Size column

df['Size'] = df['Size'].apply(convert_size)
```

```python
df['Size']
```

```python
# rename the column 
df.rename(columns={'Size':'Size_in_bytes'}, inplace=True) # rename the column 'Size' to 'Size_in_bytes'
```

```python
df.head(2)
```

```python
df['size_in_mb'] = df['Size_in_bytes'].apply(lambda x: x/(1024*1024))  # create a new column 'size_in_mb' and convert the size_in_bytes into size_in_mb
```

```python
df['size_in_kb'] = df['Size_in_bytes'].apply(lambda x: x/(1024)) # create a new column 'size_in_kb' and convert the size_in_bytes into size_in_kb
```

# Installs column

1. Remove + sign
2. Remove , sign
3. Convert the Installs column into integer

```python
df['Installs'] = df['Installs'].apply(lambda x: x.replace('+','') if '+' in str(x) else x) # replace '+' with '' in the column 'Installs'
df['Installs'] = df['Installs'].apply(lambda x: x.replace(',','') if ',' in str(x) else x) # replace ',' with '' in the column 'Installs'
df['Installs'] = df['Installs'].apply(lambda x: int(x)) # convert the column 'Installs' into integer
```

# Price Column

```python
df['Price'].value_counts() # print the value counts of the column 'Price'
```

```python
#  Let's see how many values have $ in the column 'Price'

df['Price'].str.contains('\$').sum()
```

```python
df['Price'].loc[(df['Price'].str.contains('0')) & (~df['Price'].str.contains('\$'))].value_counts().sum() # print the value counts of the column 'Price' having 0 in it but not having $ in it  
```

```python
df['Price'] = df['Price'].apply(lambda x: x.replace('$','') if '$' in str(x) else x) # replace '$' with '' in the column 'Price'
```

```python
df['Price'].value_counts() # print the value counts of the column 'Price'
```

```python
# Now convert Price column into a numeric column

df['Price'] = df['Price'].apply(lambda x: float(x)) # convert the column 'Price' into float
```
